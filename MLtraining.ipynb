```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
target_col = 'attack_cat'
```

```python
train_inputs = pd.read_parquet('train_inputs.parquet')
val_inputs = pd.read_parquet('val_inputs.parquet')
test_inputs = pd.read_parquet('test_inputs.parquet')
train_targets = pd.read_parquet('train_targets.parquet')[target_col]
val_targets = pd.read_parquet('val_targets.parquet')[target_col]
test_targets = pd.read_parquet('test_targets.parquet')[target_col]
```

```python
train_inputs
```

```python
train_inputs.sort_index(axis=1, inplace=True)
val_inputs.sort_index(axis=1,inplace=True)
test_inputs.sort_index(axis=1, inplace=True)
```

```python
model = RandomForestClassifier(n_jobs= -1 , random_state=42 , bootstrap= False)
model.fit(train_inputs, train_targets)
```

```python
pickle.dump(model,open('model.pkl','wb'))
```

```python
train_preds = model.predict(train_inputs)
accuracy_score(train_preds,train_targets)
```

```python
model.score(val_inputs,val_targets)
```

```python
model.score(test_inputs,test_targets)
```
